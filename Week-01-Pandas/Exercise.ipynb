## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 

# Select prices below 100
select_condition = df['price'] < 100
# if we print the select condition it will be a big long array of true/false.

df[select_condition].shape

# ANSWER : There are 22778 listings that ave a price less than 100.
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.

select_condition = df['neighbourhood_group'] == 'Brooklyn'

df_bk = df[select_condition]
#df_bk.head()

df_bk.shape

# ANSWER : There are 18632 listings in Brooklyn.
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

condition_1 = df['neighbourhood_group'] == 'Brooklyn'
condition_2 = df['price'] < 100

bk_low = df[(condition_1 & condition_2)]
bk_low.head()

bk_low.shape

# ANSWER : There are 10473 listings in Brooklyn with a price less than 100.
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.

names = ['Michael', 'David', 'John', 'Daniel']

condition = df.host_name.isin(names)

# Display filtered data
df[condition]

# value_counts() counts how many times each of the values in the array appear in the column
df[condition]['host_name'].value_counts()
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

condition = df['neighbourhood'] == 'Williamsburg'

# where() is similiar to an if-statement in C++
# so where the condition is true, increase price by 100, else keep the same price
df['adjusted_price'] = np.where(condition, df['price'] + 100, df['price'])

df.head(16)
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

# Counts the number of each type of room / by the total # of rooms
percentofeachroomtype = (df.room_type.value_counts() / df.room_type.value_counts().sum())
percentofeachroomtype

# ANSWER : About 45% of rooms are private and about 2% of rooms are shared.
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
gb = df.groupby('neighbourhood_group')
gb['neighbourhood_group'].count()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
df.groupby("neighbourhood_group").price.mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
df.groupby("neighbourhood_group").price.agg( ['min', 'max'] )
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
df.groupby(["neighbourhood_group", "room_type"]).price.mean()
# OR
groupby_cols = ['neighbourhood_group', 'room_type']
gb = df.groupby(groupby_cols)
gb['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
gb = df.groupby(["neighbourhood_group", "room_type"]).price.agg( ['count', 'min', 'max', 'mean', 'median', 'std'] )

#To move the 1st col_name up and make it look better
new_df = pd.DataFrame(gb).reset_index()
new_df
```

# Join and file saving.
1. Load the `prices.csv` and the `n_listings.csv`

```python
# 1. Load the `prices.csv` and the `n_listings.csv`
df_prices = pd.read_csv('data/prices.csv')
df_listings = pd.read_csv('data/n_listings.csv', sep = ';')
```

```python
# 2. Do join that keeps all the records for each table.
dfjoined = pd.merge(df_prices, df_listings, on='neighbourhood_group', how='outer')

# SAVE and EXPORT file
save_as = 'joined.csv'
dfjoined.to_csv(save_as, index=False)
#If you do this you can paste your data eleswhere
#dfjoined.to_clipboard(index=False) 

dfjoined
```

# Use the grammys.csv data for the next section of questions.

1. Who was won Album of the Year in 2016?


2. Who won Best Rap Album in 2009?


3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?

```python
df = pd.read_csv('data/grammys.csv')
df.head()
```

```python
#  1. Who was won Album of the Year in 2016?
df[(df['year'] == 2016) & 
   (df['category'] == 'Album of the Year') & 
   (df['winner'])]

# ANS : Taylor Swift won Album of the Year in 2016.
```

```python
# 2. Who won Best Rap Album in 2009?
df[(df['year'] == 2009) & 
   (df['category'] == 'Best Rap Album') & 
   (df['winner'])]

# ANS : Tha Carter III by Lil' Wayne won Best Rap Album.
```

```python
# 3. How many awards was Kendrick Lamar nominated for, and how many did he win...?

df[(df['workers'] == 'Kendrick Lamar') & 
   ((df['winner']) | (~(df['winner'])))]

# Kendrick Lamar was nominated for 6 awards, he won 4 awards.
```
